```python
# PHARMACEUTICAL CLINICAL TRIAL ANALYSIS
# Portfolio Project by Dakshil - Analyzing Drug Research Trends

print("Pharmaceutical Clinical Trial Analysis")
print("===================================")

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print("Libraries imported successfully!")

# CREATE SAMPLE CLINICAL TRIAL DATA (Realistic for portfolio)
print("Creating realistic clinical trial dataset...")

# Generate sample data
np.random.seed(42)  # For consistent results

data = {
    'trial_id': [f'CT-{i:04d}' for i in range(1, 101)],
    'drug_name': np.random.choice(['Paracetamol','Ibuprofen','Amoxicillin','Insulin','Metformin','Atorvastatin','Aspirin'], 100),
    'phase': np.random.choice(['Phase 1', 'Phase 2', 'Phase 3', 'Phase 4'], 100, p=[0.1, 0.3, 0.5, 0.1]),
    'status': np.random.choice(['Completed', 'Recruiting', 'Terminated', 'Active', 'Suspended'], 100, p=[0.3, 0.4, 0.1, 0.15, 0.05]),
    'patients': np.random.randint(50, 2000, 100),
    'success_rate': np.random.uniform(0.3, 0.95, 100),
    'country': np.random.choice(['India','USA','UK','Germany','Japan','Brazil','Australia','Canada'], 100),
    'start_date': pd.date_range('2020-03-01', periods=100, freq='D')
}

clinical_trials = pd.DataFrame(data)

print(f" Created dataset with {len(clinical_trials)} clinical trials")
print("\n FIRST 5 ROWS:")
print(clinical_trials.head())

# Basic analysis
print("\n" + "="*50)
print(" BASIC ANALYSIS")
print("="*50)

print(f"Total trials: {len(clinical_trials)}")
print(f"Trials by country:\n{clinical_trials['country'].value_counts()}")
print(f"Trials by phase:\n{clinical_trials['phase'].value_counts().sort_index()}")

# Create visualizations
print("\n CREATING VISUALIZATIONS...")

plt.figure(figsize=(15, 10))

# Chart 1: Trial status distribution
plt.subplot(2, 2, 1)
status_counts = clinical_trials['status'].value_counts()
status_counts.plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Clinical Trials by Status')
plt.ylabel('Number of Trials')
plt.xticks(rotation=45)

# Chart 2: Trials by phase
plt.subplot(2, 2, 2)
phase_counts = clinical_trials['phase'].value_counts().sort_index()
phase_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Trials by Phase')
plt.ylabel('Number of Trials')

# Chart 3: Success rates by drug
plt.subplot(2, 2, 3)
success_by_drug = clinical_trials.groupby('drug_name')['success_rate'].mean().sort_values(ascending=False)
success_by_drug.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Average Success Rate by Drug')
plt.ylabel('Success Rate')
plt.xticks(rotation=45)

# Chart 4: Patient enrollment by country
plt.subplot(2, 2, 4)
patients_by_country = clinical_trials.groupby('country')['patients'].sum().sort_values(ascending=False)
patients_by_country.plot(kind='bar', color='purple', edgecolor='black')
plt.title('Total Patient Enrollment by Country')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Advanced insights
print("\n" + "="*50)
print(" KEY INSIGHTS & BUSINESS IMPACT")
print("="*50)

print(f"1. RESEARCH FOCUS: {clinical_trials['drug_name'].value_counts().index[0]} has the most trials")
print(f"2. SUCCESS RATES: {success_by_drug.index[0]} shows highest average success ({success_by_drug.iloc[0]:.1%})")
print(f"3. GEOGRAPHICAL SPREAD: {clinical_trials['country'].value_counts().index[0]} leads in trial numbers")
print(f"4. PATIENT SCALE: Largest trial enrolled {clinical_trials['patients'].max()} patients")
print(f"5. PHASE DISTRIBUTION: {phase_counts.index[0]} trials are most common")

print("\n PHARMA INDUSTRY APPLICATIONS:")
print("   - Drug development strategy insights")
print("   - Clinical trial planning and resource allocation")
print("   - Investment decisions in drug candidates")
print("   - Global research trend analysis")

# =============================================================================
# DAKSHIL'S PERSONAL PHARMA INSIGHTS
# =============================================================================

print("\n" + "="*60)
print("DAKSHIL'S PERSONAL PHARMA INSIGHTS")
print("="*60)

# Insight 1: Most researched drug analysis
most_researched_drug = clinical_trials['drug_name'].value_counts().index[0]
most_researched_count = clinical_trials['drug_name'].value_counts().iloc[0]

print(f"1. RESEARCH FOCUS: {most_researched_drug} has {most_researched_count} trials")
print("   This suggests potential for new formulations or expanded applications")

# Insight 2: Diabetes drug analysis
diabetes_drugs = ['Metformin', 'Insulin']
for drug in diabetes_drugs:
    if drug in clinical_trials['drug_name'].values:
        drug_trials = clinical_trials[clinical_trials['drug_name'] == drug]
        print(f"2.  {drug}: {len(drug_trials)} trials, Avg success: {drug_trials['success_rate'].mean():.1%}")
        print(f"   Important for diabetes management advancements")

# Insight 3: Patient safety perspective
avg_patients = clinical_trials['patients'].mean()
large_trials = len(clinical_trials[clinical_trials['patients'] > 1000])

print(f"3. TRIAL SCALE: Average {avg_patients:.0f} patients per trial")
print(f"   {large_trials} trials have >1000 patients - crucial for safety data")

# Insight 4: Indian research contribution
india_trials = len(clinical_trials[clinical_trials['country'] == 'India'])
if india_trials > 0:
    print(f"4. ðŸ‡®ðŸ‡³ INDIA'S ROLE: {india_trials} trials conducted in India")
    print("   Shows growing clinical research capabilities")

# Insight 5: Success rate analysis
high_success = len(clinical_trials[clinical_trials['success_rate'] > 0.8])
moderate_success = len(clinical_trials[clinical_trials['success_rate'].between(0.5, 0.8)])

print(f"5. SUCCESS RATES: {high_success} trials have >80% success")
print(f"   {moderate_success} trials have 50-80% success - promising pipeline")

print("\n MY CONCLUSION AS A PHARMACY STUDENT:")
print("   This analysis demonstrates how data skills can uncover")
print("   meaningful patterns in pharmaceutical research trends.")
print("   Understanding these patterns helps in making better")
print("   healthcare decisions and resource allocation.")
```
